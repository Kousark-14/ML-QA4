# Introduction

Clustering is an unsupervised machine learning technique used to group similar data points.

In this assignment, we implement:

1. K-Means Clustering
2. DBSCAN Clustering

and compare their results using visualization.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
```

```python
X, y = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=0.60,
    random_state=42
)

print(X[:5])
```

```python
plt.scatter(X[:,0], X[:,1])

plt.title("Original Dataset")

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

kmeans_labels = kmeans.fit_predict(X)
```

```python
plt.scatter(
    X[:,0],
    X[:,1],
    c=kmeans_labels
)

plt.title("K-Means Clustering")

plt.show()
```

# Analysis

## K-Means

K-Means divided the dataset into 4 clusters.

The algorithm uses cluster centroids to assign points to clusters.

## DBSCAN

DBSCAN groups points based on density.

It can identify noise and outliers automatically.

## Observations

Both algorithms identified cluster structures.

DBSCAN is better at handling noise and irregular shapes.

K-Means performs well when clusters are spherical and clearly separated.

# Difference Between K-Means and DBSCAN

| K-Means | DBSCAN |
|----------|---------|
| Requires number of clusters (K) | Does not require K |
| Uses centroids | Uses density |
| Sensitive to outliers | Handles outliers |
| Works well for spherical clusters | Works for arbitrary shapes |

# Which Performed Better?

For this dataset, both algorithms produced good results.

However, DBSCAN is generally better when noise and outliers are present because it can automatically identify them.

K-Means is faster and simpler when the number of clusters is known.
